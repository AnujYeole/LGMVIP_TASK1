## Image To Pencil Sketch With Python 

```python
from PIL import Image
from IPython.display import display
import cv2
```

```python
image = cv2.imread('E:\\anuj.jpg', 1)
image = cv2.resize(image, (600, 600))
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
display(Image.fromarray(image))
```

```python
grey_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(grey_image))
```

```python
image_inv = cv2.bitwise_not(grey_image)
display(Image.fromarray(image_inv))
```

```python
image_smooth = cv2.GaussianBlur(image_inv, (21, 21), 0) 
display(Image.fromarray(image_smooth))
```

```python
inverted_image = 255 - grey_image
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(grey_image, inverted_blurred, scale=256.0)
display(Image.fromarray(pencil_sketch))
```
